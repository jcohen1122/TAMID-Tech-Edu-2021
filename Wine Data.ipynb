```python
import pandas as pd
```

```python
df = pd.read_csv('wine.csv')
df
```

```python
df = df[['country','points','price','province','taster_name','variety','winery']]
df
```

```python
df = df.dropna()
df
```

```python
df = df.loc[df.country.str.contains('US') & df.variety.str.contains('Pinot Gris')]
df
```

```python
#Original CSV was larger than 50 MB, so I read the shortened version into a csv file
df.to_csv('truncated_wine.csv')
```
